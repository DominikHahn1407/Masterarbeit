```python
import os

import pandas as pd
import matplotlib.pyplot as plt
import pydicom

from PIL import Image
```

```python
BASE_DIR = os.getcwd()
DATA_DIR = os.path.join(BASE_DIR, "data")
```

```python
def display_image(path):
    dicom_file = pydicom.dcmread(path, force=True)
    image = dicom_file.pixel_array
    plt.imshow(image, cmap=plt.cm.gray)
```

```python
subfolders = dict()
all_folders = [name for name in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, name))]
for name in all_folders:
    subfolders[name] = [sub_name for sub_name in os.listdir(os.path.join(DATA_DIR, name)) if os.path.isdir(os.path.join(DATA_DIR, name, sub_name))]
```

```python
print(subfolders)
```

```python
for folder_name, subfolder in subfolders.items():
    for subfolder_name in subfolder:
        for file in os.listdir(os.path.join(DATA_DIR, folder_name, subfolder_name)):
            current_file_path = os.path.join(DATA_DIR, folder_name, subfolder_name, file)
            display_image(current_file_path)
```
