```python
import os
import shutil

source_folder = "D:/Datasets/LIDC-IDRI/LIDC-IDRI-NoNodule"
target_folder="D:/Datasets/Szenario 1/NoNodule"

```

```python
# Zielordner erstellen, falls nicht vorhanden
os.makedirs(target_folder, exist_ok=True)

# Alle Unterordner durchgehen
for root, dirs, files in os.walk(source_folder):
    # Den Namen des ersten Subfolders extrahieren
    relative_path = os.path.relpath(root, source_folder)
    first_subfolder = relative_path.split(os.sep)[0] if relative_path != "." else ""

    # Nur Dateien mit der Endung `.dcm` kopieren
    for file in files:
        if file.endswith(".dcm"):
            original_path = os.path.join(root, file)
            new_file_name = f"NN_{first_subfolder}_{file}"
            destination_path = os.path.join(target_folder, new_file_name)

            # Datei kopieren
            shutil.copy2(original_path, destination_path)
            print(f"Kopiert: {original_path} -> {destination_path}")
```
